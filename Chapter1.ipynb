# Chapter 1 Summary: The Machine Learning Landscape

## üìñ Introduction to Machine Learning
This chapter introduces the fundamental concepts of Machine Learning (ML), moving beyond the sci-fi image of robots to its real-world applications like spam filters, which have been in use since the 1990s. ML is defined as the science of programming computers to learn from data, allowing a program's performance on a task to improve with experience. For example, a spam filter is trained on labeled emails (the experience) to get better at flagging spam (the task), with its success measured by accuracy (performance).

---

## ü§î Why Use Machine Learning?
ML is highly effective for problems that are too complex for traditional programming, require extensive rule-based tuning, or exist in constantly changing environments. An ML-based spam filter, for instance, automatically learns and adapts to new spam tactics, making it more robust and easier to maintain than a system built on hardcoded rules. ML is also used for **data mining**, where algorithms are inspected to uncover new insights and patterns from large datasets.

---

## ‚öôÔ∏è Types of ML Systems
ML systems can be categorized in several ways:
- **By Supervision:**
    - **Supervised Learning**: The training data includes the desired solutions, called labels. The most common tasks are **classification** (e.g., spam filtering) and **regression** (predicting a numeric value, like a house price).
    - **Unsupervised Learning**: The training data is unlabeled, and the system tries to find hidden patterns on its own. Common tasks include **clustering**, **anomaly detection**, and **dimensionality reduction**.
    - **Semisupervised Learning**: A mix of labeled and unlabeled data is used.
    - **Reinforcement Learning**: An "agent" learns by performing actions and receiving rewards or penalties, discovering the best strategy (policy) over time.
- **By Learning Style:**
    - **Batch Learning**: The system is trained using all available data at once and cannot learn incrementally. To adapt, it must be retrained from scratch on the full dataset.
    - **Online Learning**: The system learns incrementally by being fed data sequentially, allowing it to adapt to new data on the fly.
- **By Generalization Method:**
    - **Instance-Based Learning**: The system learns examples by heart and generalizes to new cases by comparing them using a similarity measure.
    - **Model-Based Learning**: The system builds a predictive model from the data and uses that model to make predictions.

---

## ‚ö†Ô∏è Main Challenges & Solutions
The biggest hurdles in ML often relate to "bad data" or a "bad algorithm."
- **Data Challenges**:
    - **Insufficient Data**: Most algorithms need thousands or even millions of examples to perform well.
    - **Nonrepresentative Data**: The training data must be representative of the cases you want to generalize to, otherwise the model will be biased.
    - **Poor-Quality Data**: Errors, outliers, and noise make it harder for the system to detect underlying patterns.
    - **Irrelevant Features**: The principle of "garbage in, garbage out" applies; a good set of features is critical.
- **Algorithm Challenges**:
    - **Overfitting**: The model is too complex and performs well on the training data but fails to generalize to new instances. Solutions include simplifying the model, gathering more data, or cleaning the existing data.
    - **Underfitting**: The model is too simple to learn the underlying structure of the data. Solutions include selecting a more powerful model or engineering better features.

---

## ‚úÖ Testing and Validating
To know how well a model will perform, you split your data into a **training set** and a **test set**. You train the model on the training data and evaluate its **generalization error** (performance on new cases) using the test set. If the training error is low but the generalization error is high, the model is overfitting. A **validation set** is often held out from the training data to tune hyperparameters and select the best model without touching the test set.

## üíª Kode Python - Implementasi Example 1-1

Berikut adalah kode lengkap yang mengilustrasikan konsep-konsep dari Chapter 1, seperti model-based learning, visualisasi, dan perbandingan model. Setiap cell dapat dijalankan secara berurutan.

```python
# Cell 1: Import Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

print("Libraries berhasil diimport!")
```

```python
# Cell 2: Membuat Data
data = {
    'Country': ['Hungary', 'Korea', 'France', 'Australia', 'United States',
                'New Zealand', 'Norway', 'Sweden', 'Switzerland', 'Denmark',
                'Netherlands', 'Canada', 'Austria', 'United Kingdom', 'Germany'],
    'GDP per capita': [12240, 27195, 37675, 50962, 55805,
                     38967, 74822, 49866, 62882, 53579,
                     50013, 43331, 45436, 40412, 44184],
    'Life satisfaction': [4.9, 5.8, 6.5, 7.3, 7.2,
                          7.3, 7.4, 7.2, 7.5, 7.5,
                          7.4, 7.3, 7.0, 6.9, 6.7]
}

country_stats = pd.DataFrame(data)
print("Data berhasil dibuat!")
print("\nPreview data:")
print(country_stats.head())
```

```python
# Cell 3: Persiapan Data
X = country_stats[['GDP per capita']].values
y = country_stats[['Life satisfaction']].values

print(f"Shape X (features): {X.shape}")
print(f"Shape y (target): {y.shape}")
```

```python
# Cell 4: Visualisasi Data
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.8, s=100, color='steelblue')
plt.xlabel('GDP per capita (USD)')
plt.ylabel('Life Satisfaction')
plt.title('Hubungan antara GDP per Capita dan Life Satisfaction')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Cell 5: Linear Regression Model
lin_model = LinearRegression()
lin_model.fit(X, y)

print("Model Linear Regression telah dilatih!")
```

```python
# Cell 6: Prediksi untuk Cyprus
cyprus_gdp = np.array([[22587]])
cyprus_prediction = lin_model.predict(cyprus_gdp)

print(f"Prediksi Life Satisfaction untuk Cyprus (GDP ${cyprus_gdp[0][0]:,}): {cyprus_prediction[0][0]:.2f}")
```

```python
# Cell 7: Visualisasi Model Linear
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Training data')
X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_plot = lin_model.predict(X_plot)
plt.plot(X_plot, y_plot, 'r-', label='Linear Model')
plt.scatter(cyprus_gdp, cyprus_prediction, color='green', s=200, marker='*',
            label=f'Cyprus prediction: {cyprus_prediction[0][0]:.2f}')
plt.xlabel('GDP per capita (USD)')
plt.ylabel('Life Satisfaction')
plt.title('Linear Regression Model')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Cell 8: k-Nearest Neighbors Model
knn_model = KNeighborsRegressor(n_neighbors=3)
knn_model.fit(X, y)
cyprus_prediction_knn = knn_model.predict(cyprus_gdp)

print(f"Prediksi k-NN (k=3) untuk Cyprus: {cyprus_prediction_knn[0][0]:.2f}")
```

```python
# Cell 9: Ilustrasi Overfitting
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

poly_model = Pipeline([
    ('poly_features', PolynomialFeatures(degree=10, include_bias=False)),
    ('lin_reg', LinearRegression())
])
poly_model.fit(X, y)

plt.figure(figsize=(12, 6))
plt.scatter(X, y, label='Training data')
X_plot = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)
plt.plot(X_plot, lin_model.predict(X_plot), 'g-', label='Linear (mungkin underfitting)')
plt.plot(X_plot, poly_model.predict(X_plot), 'r-', label='Polynomial degree 10 (overfitting)')
plt.title('Ilustrasi Overfitting vs Underfitting')
plt.legend()
plt.grid(True)
plt.ylim(4, 8)
plt.show()
```

```python
# Cell 10: Evaluasi Model pada Training Data
from sklearn.metrics import mean_squared_error

y_pred_lin = lin_model.predict(X)
rmse_lin = np.sqrt(mean_squared_error(y, y_pred_lin))

y_pred_poly = poly_model.predict(X)
rmse_poly = np.sqrt(mean_squared_error(y, y_pred_poly))

print("EVALUASI MODEL (pada Training Data)")
print(f"RMSE Linear Regression: {rmse_lin:.4f}")
print(f"RMSE Polynomial (d=10): {rmse_poly:.4f}")
print("\nNote: Model polynomial memiliki error lebih kecil di training data,")
print("tetapi ini adalah tanda overfitting dan akan perform buruk pada data baru.")
```
